```python
# Charger collect.txt
with open("collect.txt", "r", encoding="utf-8") as file:
    corpus = file.read()

print(corpus[:500])  # Afficher un aperçu du texte (500 premiers caractères)
```

```python
# Charger stoplist.txt
with open("stoplist.txt", "r", encoding="utf-8") as file:
    stop_words = file.read().splitlines()

print(stop_words[:20])  # Aperçu des stop words
```

```python
import os

# Parcourir le dossier Collection_TIME
folder_path = "Collection_TIME"
files = os.listdir(folder_path)

corpus_files = []
for filename in files:
    file_path = os.path.join(folder_path, filename)
    with open(file_path, "r", encoding="utf-8") as file:
        corpus_files.append(file.read())

print(f"Nombre de fichiers chargés : {len(corpus_files)}")
print(corpus_files[0][:500])  # Afficher un extrait du premier fichier
```

```python
# Combiner le texte de collect.txt et les fichiers de Collection_TIME
all_text = corpus + "\n".join(corpus_files)
print(f"Nombre total de caractères dans le corpus : {len(all_text)}")
```

```python
import re
# Alternative à word_tokenize
def preprocess(text, stop_words):
    text = text.lower()
    text = re.sub(r'\W', ' ', text)
    tokens = text.split()  # Alternative simple
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)
```

```python
# Prétraiter tout le corpus
cleaned_text = preprocess(all_text, stop_words)
print(cleaned_text[:500])  # Aperçu du texte nettoyé
```

```python
# Sauvegarder le corpus nettoyé
with open("cleaned_corpus.txt", "w", encoding="utf-8") as file:
    file.write(cleaned_text)
```

```python
import numpy as np
from collections import Counter
from math import log

# Fonction pour calculer TF-IDF manuellement
def compute_tfidf(corpus, max_features=1000):
    # Tokenisation
    words = corpus.split()
    
    # Calcul de TF
    term_counts = Counter(words)
    total_terms = len(words)
    tf = {term: count / total_terms for term, count in term_counts.items()}
    
    # Calcul de DF
    df = Counter()
    df.update(term_counts.keys())  # Chaque mot apparaît dans ce document
    
    # Calcul de IDF
    total_docs = 1  # Un seul document dans cet exemple
    idf = {term: log((total_docs + 1) / (df[term] + 1)) + 1 for term in term_counts}
    
    # Calcul TF-IDF
    tfidf = {term: tf[term] * idf[term] for term in term_counts}
    
    # Sélection des max_features mots les plus fréquents
    sorted_tfidf = sorted(tfidf.items(), key=lambda x: x[1], reverse=True)[:max_features]
    feature_names = [term for term, _ in sorted_tfidf]
    values = [tfidf[term] for term in feature_names]
    
    return np.array(values), feature_names

# Charger le texte nettoyé
with open("cleaned_corpus.txt", "r", encoding="utf-8") as file:
    cleaned_text = file.read()

# Calculer TF-IDF
tfidf_values, feature_names = compute_tfidf(cleaned_text)

print("TF-IDF Shape :", len(tfidf_values))
print("Top 10 termes :", feature_names[:10])
```

```python
from gensim.models import TfidfModel
from gensim.corpora import Dictionary

# Charger le texte nettoyé
with open("cleaned_corpus.txt", "r", encoding="utf-8") as file:
    cleaned_text = file.read()

# Préparer le corpus
texts = [cleaned_text.split()]  # Tokeniser par mots
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Calculer TF-IDF
tfidf_model = TfidfModel(corpus)
tfidf_corpus = tfidf_model[corpus]

# Convertir en matrice dense
tfidf_matrix = gensim.matutils.corpus2dense(tfidf_corpus, num_terms=len(dictionary)).T

print("TF-IDF Shape :", tfidf_matrix.shape)
print("Top 10 termes :", list(dictionary.token2id.keys())[:10])
```
